```python
df_3=df.groupby(["Grand Prix", "Year"]).agg("mean", numeric_only=True).reset_index()
df["Average Lap Pitted"]=df.groupby(["Grand Prix", "Year"])["Lap"].transform("mean")
df
```

```python
df_early=df[df["Time Pitted"]=="Early"]
df["Q1 Lap Pitted"]=df_early.groupby(["Grand Prix", "Year"])["Lap"].transform("mean")
df_late=df[df["Time Pitted"]=="Late"]
df["Q3 Lap Pitted"]=df_late.groupby(["Grand Prix", "Year"])["Lap"].transform("mean")
df
```

```python
for i in range (len(df)):
    if df.loc[i,"Lap"] <= df.loc[i,"Q1 Lap Pitted"]:
        df.loc[i,"Time Pitted"]="Very Early"
    elif (df.loc[i,"Lap"] <= df.loc[i, "Average Lap Pitted"]) & (df.loc[i,"Lap"]>=df.loc[i, "Q1 Lap Pitted"]):
        df.loc[i, "Time Pitted"]="Early"
    elif (df.loc[i,"Lap"] <= df.loc[i, "Q3 Lap Pitted"]) & (df.loc[i,"Lap"]>=df.loc[i, "Average Lap Pitted"]):
        df.loc[i, "Time Pitted"]="Late"
    else:
        df.loc[i,"Time Pitted"]="Very Late"
df
```
