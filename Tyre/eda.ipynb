```python
# Load the first 42 rows of the CSV file
file_path = "2019-2024 f1 tyre data(Sheet1).csv"  # Update with your file path if needed
data = pd.read_csv(file_path).iloc[:42]

# Generate finishing position based on consecutive driver names
data['Position'] = (data['Driver'] != data['Driver'].shift()).cumsum()

# Plotting: Tyre Type vs. Finishing Place
plt.figure(figsize=(10, 6))
plt.scatter(data['Position'], data['Tyres'], color='b', marker='o')

# Set plot labels and title
plt.xlabel("Finishing Place")
plt.ylabel("Tyre Type")
plt.title("Tyre Type by Finishing Place (First 42 Rows)")
plt.xticks(range(1, data['Position'].nunique() + 1))  # Unique positions for ticks
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
